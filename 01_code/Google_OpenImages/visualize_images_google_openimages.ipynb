```python
# load libraries
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
```

```python
# load csv file with box annotations
person_annotations = pd.read_csv('../../02_data/Google_OpenImages/filtered_person_annotations.csv')
```

```python
image_folder = '../../02_data/Google_OpenImages/person_images_small/' # specify image folder
image_id = '009e94e6f82287e6'  # Replace with your actual image ID
image_file = os.path.join(image_folder, "{}.jpg".format(image_id))  # specify image
image_boxes = person_annotations[person_annotations['ImageID'] == image_id] # get only relevant rows
```

```python
# show the image
img = Image.open(image_file)
plt.figure(figsize=(12, 8))
plt.axis('off')
plt.imshow(img)
```

```python
# show the image with bounding boxes
draw = ImageDraw.Draw(img)

for index, row in image_boxes.iterrows():
    x_min = row['XMin'] * img.width
    y_min = row['YMin'] * img.height
    x_max = row['XMax'] * img.width
    y_max = row['YMax'] * img.height
    
    draw.rectangle([x_min, y_min, x_max, y_max], outline='red', width=2)

plt.figure(figsize=(12, 8))
plt.imshow(img)
plt.axis('off')
plt.show()
```
